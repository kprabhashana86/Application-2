```python
#Importing Libraries
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Reading Excel file while not viewing the first row.
df = pd.read_excel('data\Moving annual rent by suburb - September quarter 2022.xlsx', sheet_name='All properties',
            header=1)
# Renaming the Columns.
df = df.rename(columns={'All properties': 'Region', "Unnamed: 1": "Suburb"}, index={'ONE': 'Row_1'})


# Renaming the unnamed columns.
def repeat_cols(columns):
    last_non_unnamed = '' if 'unnamed' in columns[0].lower() else columns[0]
    new_columns = []
    for column in columns:
        if 'unnamed' in column.lower():
            new_column = last_non_unnamed
        else:
            new_column = column
            last_non_unnamed = new_column
        new_columns.append(new_column)
    return new_columns

df.columns = repeat_cols(df.columns)

df.columns = [f'{a} {b}' if pd.notna(b) else a for a, b in zip(df.columns, df.iloc[0])]
df = df.iloc[1:].reset_index(drop=True)


 #Renaming the rows with filling.
df.fillna(method='ffill', inplace=True)


# Viewing the dataframe named df.
df.head()
```

```python
# Dataset 3 'Group Total'
df3 = df.loc[df['Suburb'] == 'Group Total']

# Remove Group total Rows from main dataframe.
df = df[df.Suburb != 'Group Total']
```

## 4) Test the following hypothesis
### i) Is Distance to Melbourne is a factor affecting the rental price

```python
# Joining Sep 20220 Count column from dataframe 2 to dataframe 1
df1 = df1.join(df2[['Sep 2022 Count']])
```

```python
# Get dataset with region, Suburb, Sep 2022 Count and distance columns
df_test = df1[['Region','Suburb','Sep 2022 Count','distance']]
df_test.head()
```

```python
# get data with following condition
df_con_1 = df_test.query("distance < 5")
df_con_2 = df_test.query("5 < distance < 12")
df_con_3 = df_test.query(" 12 < distance ")
```

```python
# importing stats
from scipy  import stats
```

```python
# one-way ANOVA test is used 
f_val, p_val = stats.f_oneway(df_con_1['distance'],df_con_2['distance'],
                              df_con_3['distance'])
# If p value is > 0.05 then we can assume that there is connection  between two
print("ANOVA results: F=",f_val, ",P=",p_val)
print('Distance to Melbourne is a factor affecting the rental price')
```

### Since P value is > 0.05 Distance to Melbourne is a factor affecting the rental price

### ii) The direction from Melbourne is a factor affecting the rental price

```python
# Getting direction between melbourne and Suburbs
x1 = -37.8136
y1 = 144.9631

A = np.degrees(np.arctan((df1['Longitude']-y1)/(df1['Latitude']-x1)))
```

```python
df1['Angle'] = A
```

```python
# Angle in degrees column is named as RAngle
df1['RAngle'] = (
    np.select(
        condlist=[(df1['Longitude'] > 144.9631) & (df1['Latitude'] < -37.8136), (df1['Longitude'] < 144.9631) & (df1['Latitude'] < -37.8136),(df1['Longitude'] < 144.9631) & (df1['Latitude'] > -37.8136)], 
        choicelist=[180-df1['Angle'] ,180+df1['Angle'],360-df1['Angle']],
        default=df1['Angle']))
df1
```

```python
# Get dataset with region, Suburb, Sep 2022 Median and RAngle columns
df_test1 = df1[['Region','Suburb','Sep 2022 Median','RAngle']]
```

```python
df_test1
```

```python
df_test1.info()
```

```python
## get data with following condition

dft1 = df_test1.query("0< RAngle < 120")
dft2 = df_test1.query(" 120 < RAngle < 240 ")
dft3 = df_test1.query(" 240 < RAngle < 360 ")
```

```python
# one-way ANOVA test is used 
f_val1, p_val1 = stats.f_oneway(dft1['RAngle'],dft3['RAngle'],dft3['RAngle'])#,dft4['RAngle'])
print("ANOVA results: F=",f_val1, ",P=",p_val1)
print('Since P value is < 0.05 The direction from Melbourne is not a factor affecting the rental price')
```
